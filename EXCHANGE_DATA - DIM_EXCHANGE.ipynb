```python
df = pd.read_csv('exchange_wallets_clean.csv')
df
```

```python
df.exchange_name.nunique()
```

```python
# the rows for distinct wallet_id
df_distinct = df.drop_duplicates(subset=['exchange_name'])

# give the distinct wallet_id a unique id starting from 1
df_distinct['exchange_id'] = range(1, 1+len(df_distinct))
df_distinct
```

```python
# find the wallet_id_2 for each wallet_id in the original df
df = pd.merge(df, df_distinct[['exchange_name', 'exchange_id']], on='exchange_name', how='left')
df
```

```python
df_ticker = pd.read_csv('DIM_TICKER.csv')
df_ticker
```

```python
# add ticker_id from df_ticker to df
df = pd.merge(df, df_ticker[['symbol', 'TICKER_ID']], left_on='ticker_name' ,right_on='symbol', how='left')
df
```

# EXCHANGE_DATA TABLE CREATION

```python
df.drop(columns=['ticker_name', 'exchange_name', 'symbol'], inplace=True)

# bring exchange_id and ticker_id to the front
cols = df.columns.tolist()
cols = cols[-2:] + cols[:-2]
df = df[cols]
df
```

```python
df.rename(columns = {'TICKER_ID': 'ticker_id'}, inplace=True)
df
```

```python
df.to_csv('EXCHANGE_DATA.csv', index=False)
```
