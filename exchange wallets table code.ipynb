```python
df = pd.read_csv('exchange wallets.csv')
df
```

```python
df = df.fillna(0)
```

```python
# remove the number strings in the ticker name column
df['ticker_name'] = df['ticker_name'].str.replace(r'\d+', '')

# remove the spaces in the ticker name column
df['ticker_name'] = df['ticker_name'].str.replace(' ', '')
```

```python
df
```

```python
df.info()
```

```python
lst_col = ['ticker_exchange_tokenbalance', 'ticker_exchange_selection1', 'ticker_exchange_sevendaychange', 'ticker_exchange_threedaychange']

for col in lst_col:
    try:
        # get rid of the letters in the columns
        df[col] = df[col].str.replace(r'[a-zA-Z]+', '')

        # get rid of the spaces in the columns
        df[col] = df[col].str.replace(' ', '')

        # get rid of the commas in the columns
        df[col] = df[col].str.replace(',', '')

        # turn the columns into floats
        df[col] = df[col].astype(float)

        # change NaN values to 0
        df[col] = df[col].fillna(0)
    except Exception as e:
        # print the row that caused the error
        pass
```

```python
# replace the empty strings with 0
df['ticker_exchange_threedaychange'] = df['ticker_exchange_threedaychange'].replace('', 0)
# replace -- with 0
df['ticker_exchange_threedaychange'] = df['ticker_exchange_threedaychange'].replace('--', 0)
df['ticker_exchange_threedaychange'] = df['ticker_exchange_threedaychange'].astype(float)


# Same with sevendaychange column
df['ticker_exchange_sevendaychange'] = df['ticker_exchange_sevendaychange'].replace('', 0)
df['ticker_exchange_sevendaychange'] = df['ticker_exchange_sevendaychange'].replace('--', 0)
df['ticker_exchange_sevendaychange'] = df['ticker_exchange_sevendaychange'].astype(float)
```

```python
df.columns = ['ticker_name', 'ticker_url', 'exchange_name', 'exchange_token_balance (cryptocurrency)', 'exchange_value', 'selection1 (cryptocurrency)',
                    'seven_day_change (cryptocurrency)', 'three_day_change (cryptocurrency)']
    
    # reset the index
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('exchange wallets all.csv', index=False)
```
