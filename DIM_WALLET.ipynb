```python
df = pd.read_csv('rich_list_clean.csv')
df
```

```python
df_ticker = pd.read_csv('DIM_TICKER.csv')
df_ticker
```

```python
# the rows for distinct wallet_id
df_distinct = df.drop_duplicates(subset=['wallet_id'])

# give the distinct wallet_id a unique id starting from 100000
df_distinct['wallet_id_2'] = range(100001, 100001+len(df_distinct))
df_distinct
```

```python
# find the wallet_id_2 for each wallet_id in the original df
df = pd.merge(df, df_distinct[['wallet_id', 'wallet_id_2']], on='wallet_id', how='left')
df
```

```python
# add ticker_id from df_ticker to df
df = pd.merge(df, df_ticker[['symbol', 'TICKER_ID']], left_on='ticker' ,right_on='symbol', how='left')
df
```

```python
df.drop(columns=['rank', 'ticker', 'currency_quantity (cryptocurrency)', 
                 'fraction_of_total (%)', 'seven_day_change (cryptocurrency)'], inplace=True)
```

```python
df.drop(columns=['symbol'], inplace=True)
```

```python
df.columns = ['wallet_guid', 'ticker_id', 'wallet_id']
df
```
