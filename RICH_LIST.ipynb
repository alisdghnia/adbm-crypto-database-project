```python
df = pd.read_csv('rich_list_clean.csv')
df
```

```python
df_wallet = pd.read_csv('DIM_WALLET.csv')
df_wallet
```

```python
# iterate over rows and add wallet_id and ticker_id to df
for index, row in df.iterrows():
    wallet_id = df_wallet[df_wallet['wallet_guid'] == row['wallet_id']]['wallet_id'].values[0]
    ticker_id = df_wallet[df_wallet['wallet_guid'] == row['wallet_id']]['ticker_id'].values[0]
    df.at[index, 'wallet_id'] = wallet_id
    df.at[index, 'ticker_id'] = ticker_id

df
```

```python
df.drop(columns=['ticker', 'rank', 'ticker_id'], inplace=True)
df
```

```python
df.columns = ['wallet_id', 'currency_quantity (cryptocurrency)', 'fraction_of_total (%)', 'seven_day_change (cryptocurrency)']
df
```

```python
df.to_csv('RICH_LIST.csv', index=False)
```
