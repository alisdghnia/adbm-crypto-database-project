```python
df = pd.read_csv('Crypto Data.csv')
df
```

```python
df.columns
```

```python
df.columns = ['name', 'symbol', 'rank', 'price (USD)', 'oneday (%)', 'oneweek (%)', 'marketcap (USD)', 'volume (USD)', 'circulating_supply (Crypto)', 'crypto_name_symbol', 'full_marketcap (USD)', 'max_supply (Crypto)',
              'oneday_low (USD)', 'oneday_high (USD)', 'volume_marketcap (%)', 'market_dominance (%)', 'onemonth_ROI (%)', 'price_change (USD)']
```

```python
df = df.drop('crypto_name_symbol', axis=1)
df
```

```python
# df.to_csv('Cleaned Crypto Data.csv', index=False)
```

```python
df.info()
```

```python
# get rid of $ and , in all columns
for i in df.columns:
    try:    
        df[i] = df[i].str.replace('$', '')
        df[i] = df[i].str.replace(',', '')
        df[i] = df[i].str.replace('%', '')
        df[i] = df[i].str.replace(' ', '')
        # df[i] = df[i].str.replace('N/A', '0')
    except Exception:
        pass
df
```

```python
# get rid of numbers in symbol column
df['symbol'] = df['symbol'].str.replace('[0-9]', '')
df
```

```python
# get rid of letters in circulating_supply and max_supply columns
df['circulating_supply (Crypto)'] = df['circulating_supply (Crypto)'].str.replace('[a-zA-Z]', '')
df['max_supply (Crypto)'] = df['max_supply (Crypto)'].str.replace('[a-zA-Z]', '')
df
```

```python
# turn all columns into floats except for name, symbol, and rank with 2 decimal places
for i in df.columns:
    try:
        df[i] = df[i].astype(float)
        df[i] = df[i].round(2)
    except Exception:
        pass

df['rank'] = df['rank'].astype(int)

df
```
