```python
df = pd.read_csv('Metadata 1838 Whitepapers copy.csv')
df
```

```python
# get rid of - Whitepaper.pdf in the Name column
df['Name'] = df['Name'].str.replace(' - Whitepaper.pdf', '')
df
```

```python
df_ticker = pd.read_csv('DIM_TICKER.csv')
name_lst = list(df_ticker.name)
name_lst
```

```python
# rows in df that are in the name_lst
df_1 = df[df['Name'].isin(name_lst)]
```

```python
df_1.reset_index(drop=True, inplace=True)
```

```python
df_final = pd.merge(df_ticker, df_1, how='left', left_on='name', right_on='Name')
```

```python
df_final.drop(columns=['symbol', 'name', 'rank', 'Unnamed: 0'], inplace=True)
df_final
```

```python
df_final.dropna(inplace=True)
df_final
```

```python
# add whitepaper_id column starting from 10000001
df_final['whitepaper_id'] = range(10000001, 10000001+len(df_final))
df_final
```

```python
# add two columns with the names of title and release_date and give them null values
df_final['title'] = None
df_final['release_date'] = None
df_final
```

```python
# bring the last 3 columns to the front
cols = df_final.columns.tolist()
cols = cols[-3:] + cols[:-3]
df_final = df_final[cols]
df_final
```

```python
df_final.drop(columns=['Name'], inplace=True)
```
